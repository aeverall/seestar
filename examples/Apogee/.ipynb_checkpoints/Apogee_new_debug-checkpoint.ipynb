```python
sub = pd.read_csv('PATH/Apogee_new/Apogee_new_unionobssolution.csv')
```

### Field ID 1

```python
fig = plt.figure(figsize=(20,20))
plt.subplot(projection='mollweide')

sub1 = sub[sub.fieldID ==1].copy()
sub1.glon[sub1.glon>np.pi] -= 2*np.pi

plt.scatter(sub1.glon, sub1.glat)
```

### No field assignment

```python
sub2 = sub[(sub.union==0)&(sub.fieldID != 1)].copy()
```

```python
fig = plt.figure(figsize=(20,20))
plt.subplot(projection='mollweide')

arr = np.array(sub2.glon)
arr[arr>np.pi] -= 2*np.pi
sub2.glon = arr

plt.scatter(sub2.glon, sub2.glat)
```

```python
np.unique(sub2.fieldID)
```

```python
pointings = pd.read_csv(fileinfo.field_path)
pointings.glon[pointings.glon>np.pi] -= 2*np.pi
pointings.set_index('fieldID', inplace=True)
```

```python
field = 5128

subfield = sub[sub.fieldID == field].copy()
pointfield = pointings[pointings.fieldID == field].copy()

arr = np.array(subfield.glon)
arr[arr>np.pi] -= 2*np.pi
subfield.glon = arr
arr = np.array(pointfield.glon)
arr[arr>np.pi] -= 2*np.pi
pointfield.glon = arr

plt.scatter(subfield[subfield.union>0].glon, subfield[subfield.union>0].glat, c='g', s=10)
plt.scatter(subfield[subfield.union==0].glon, subfield[subfield.union==0].glat, c='b', s=10)
plt.scatter(pointfield.glon, pointfield.glat, c='r', s=100)
```

```python
pointings[pointings.halfangle<0.026]
```

```python
field = 4361

photo = pd.read_csv(os.path.join(fileinfo.photo_path, str(field))+'.csv')
pointf = pointings.loc[field]
surveyf = survey[survey.fieldID == field].copy()


photo.glon[photo.glon>np.pi] -= 2*np.pi
surveyf.glon[surveyf.glon>np.pi] -= 2*np.pi

con = np.pi/180
patch = matplotlib.patches.Ellipse((pointf.glon, pointf.glat), 
                                   2*pointf.halfangle/np.cos(pointf.glat), 2*pointf.halfangle, color='r', fill=True, alpha=0.1)

fig = plt.figure(figsize=(15, 18))
ax = plt.subplot()

plt.scatter(photo.glon, photo.glat, label='Field 3',
            marker='o', c='b', s=0.1)
plt.scatter(surveyf.glon, surveyf.glat, label='Field2',
            marker='o', c='none', edgecolors='g', s=10, linewidths=3)
#plt.scatter(df23.glon, df23.glat, c='k', s=0.1)

ax.add_artist(patch)

plt.rc('mathtext', fontset='cm')
plt.rc('font', family='serif')
plt.xlabel(r'$l$', fontsize=40, style='italic')
plt.ylabel(r'$b$', fontsize=40, style='italic')
plt.tick_params(axis='both', labelsize=30)

plt.legend(fontsize=25, frameon=True, markerscale=2)
```
