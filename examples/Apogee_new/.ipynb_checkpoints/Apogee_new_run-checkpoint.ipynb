```python
# Plots for each of the 4 quantities

def HistogramValues(cat, label, nbin=20, rng=(0,1),
                   lweight='', weighted=False):   

    if  weighted: 
        density, val = np.histogram(cat[label], weights=1/cat[lweight], bins=nbin, range=rng)
        N = np.sum(1/cat[lweight])
    else: 
        density, val = np.histogram(cat[label], bins=nbin, range=rng)
        N = float(len(cat))
        
    centers = ArrayMechanics.BoundaryToCentre(val)
    
    return density/N, centers
```

```python
fig = plt.figure(figsize=(20, 20))

sub1 = sub[sub.union>1e-1]

nbin=50

ax1 = plt.subplot(111)
age_d, age = HistogramValues(sub1, 's', nbin=nbin, rng=(0, 10))
_=plt.plot(age, age_d, label='full', linewidth=2.)
age_d, age = HistogramValues(sub1, 's', nbin=nbin, rng=(0, 10), weighted=True, lweight='union')
_=plt.plot(age, age_d, label='weighted', linewidth=2.)
ax1.set_xlabel(r"$\tau$")
ax1.set_ylabel('Normalised count')
```
